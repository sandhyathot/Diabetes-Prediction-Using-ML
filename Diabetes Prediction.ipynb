```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['diabetes'],axis=1)
y=df.diabetes.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
print('Test Accuracy:{:.2f}%'.format(lr.score(x_test,y_test)*100))
```

```python
#KNeighbors
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=25)
knc.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(knc.score(x_test,y_test)*100))
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(gb.score(x_test,y_test)*100))
```

```python
#SVC
from sklearn.svm import SVC
svc=SVC(random_state=1)
svc.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(svc.score(x_test,y_test)*100))
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(n_estimators=500,random_state=1)
rc.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(rc.score(x_test,y_test)*100))
```

```python
#xgboost
import xgboost
xgb=xgboost.XGBClassifier()
xgb.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(xgb.score(x_test,y_test)*100))
```

```python
#Logistic Regression model Accuracy: 78.12%
#KNeighbors model Accuracy: 77.08%
#Naive Bayes model Accuracy: 78.12%
#SVC model Accuracy: 64.06%
#Random Forest Classifier model Accuracy: 81.77%
#xgboost model Accuracy: 76.04%
```
